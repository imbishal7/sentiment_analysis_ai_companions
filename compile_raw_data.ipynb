```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import os
```

```python
unique_review_class = 'RHo1pe'
date_class = 'bp9Aid'
rating_class = 'iXRFPc'
review_class = 'h3YV2d'

df_columns = ['app', 'date', 'review', 'rating']
```

```python
appss = ["replika", "snapchat",'chatgpt', 'googleassistant']

for app_name in appss:

  file_path = f"/content/drive/MyDrive/Study/Spring 24/HON 112/FinalProject/raw/{app_name}/"
  save_path = f"/content/drive/MyDrive/Study/Spring 24/HON 112/FinalProject/data/{app_name}_reviews.csv"
  result = []

  def html_to_csv(file_name):
    full_path = file_path + file_name
    print(f'Now making soup for {file_name}')
    soup = BeautifulSoup(open(full_path), 'html.parser')
    unique_revs = soup.find_all('div', {'class': unique_review_class})

    for i in unique_revs:
      try:
        date = i.find('span', {'class': date_class}).text
      except:
        date = 'NA'
      try:
        rating = i.find('div', {'class': rating_class})['aria-label']
      except:
        rating = 'NA'
      try:
        review = i.find('div', {'class': review_class}).text
      except:
        review = 'NA'

      result.append([app_name, date, review, rating])

  for i in os.listdir(file_path):
    html_to_csv(i)

  df = pd.DataFrame(result, columns = df_columns)
  df['rating'] = df['rating'].apply(lambda x: x.split(' ')[1])

  df.to_csv(save_path)
```
